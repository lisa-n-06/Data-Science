```python
#salary dataset
import numpy as np 
import pandas as pd
data=np.array([18000,20000,22000,25000,27000,30000,150000])
data
```

```python
#outliers 

data.mean()
```

```python
from scipy.stats import skew
```

```python
#skew 
skew(data)
```

```python
import pandas as pd
d=pd.DataFrame({"values":data})
d
```

```python
#right skew
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(d,kde=True)
plt.show()
```
