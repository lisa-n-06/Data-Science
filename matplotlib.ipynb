```python
#matplotlib
import pandas as pd
d=pd.read_csv(r"C:\Users\Mohan\Downloads\StudentsPerformance.csv")
d
```

```python
import matplotlib.pyplot as plt
```

```python
d.info()
```

```python
d.groupby("gender")[["math score","reading score","writing score"]].mean().plot(kind="bar")
plt.title("gender vs score")
plt.xlabel("gender")
plt.ylabel("score")
plt.show()
```

```python
a=["juice","snacks","fastfood"]
b=[520,629,125]
plt.bar(a,b)
plt.title("Variety vs Having Percentage")
plt.xlabel("Variety of food")
plt.ylabel("percentage of having")
plt.show()
```

```python
plt.hist(d["math score"],bins=20)
plt.show()
```

```python
plt.scatter(d["math score"],d["reading score"])
plt.show()
```

```python
plt.plot(d["math score"],d["reading score"])
plt.show()
```

```python
avg_s=d[['math score',"reading score","writing score"]].mean()
avg_s


x=avg_s.index
y=avg_s.values

plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("values")
plt.ylabel("mean") 
plt.show()
```
